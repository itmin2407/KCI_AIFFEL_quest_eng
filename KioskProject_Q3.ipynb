```python
class Kiosk:
    def __init__(self):
        # 2) 메뉴와 가격을 다차원 리스트로 변경
        self.menu = [
            ['americano', 2000],
            ['latte', 3000],
            ['mocha', 3000],
            ['yuza_tea', 2500],
            ['green_tea', 2500],
            ['choco_latte', 3000]
        ]
        self.order_menu = []
        self.order_price = []

    # 3) 메뉴 출력 enumerate() 함수 활용
    def menu_print(self):
        for i, item in enumerate(self.menu, 1):
            print(i, item[0], ' : ', item[1], '원')

    # 1) 주문 메서드
    def menu_select(self):
        print()
        n = int(input("음료 번호를 입력하세요 : "))
        self.price_sum = self.menu[n-1][1]
        print(self.menu[n-1][0], ':', self.menu[n-1][1], '원')

        while n != 0:
            t = 0
            while t != 1 and t != 2:
                t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                if t == 1:
                    self.temp = "HOT"
                elif t == 2:
                    self.temp = "ICE"
                else:
                    print("1과 2 중 하나를 입력하세요.\n")
            print('추가 주문 음료', self.temp, self.menu[n-1][0], ':', self.menu[n-1][1], '원')

            # 1) 추가 주문 메서드
            while n != 0:
                print()
                n = int(input('추가 주문은 음료 번호를,지불은 0을 누르세요:'))
                if n > 0 and n < len(self.menu) + 1:
                    self.order_price.append(self.menu[n-1][1])
                    self.order_menu.append(self.temp + ' ' + self.menu[n-1][0])
                    self.price_sum += self.menu[n-1][1]

                    print('추가 주문 음료', self.temp, self.menu[n-1][0], ' : ', self.menu[n-1][1], '원\n', '합계 : ', self.price_sum, '원')
                else:
                    if n == 0:
                        print("주문이 완료 되었습니다")
                        print(self.order_menu, self.order_price)
                    else:
                        print("없는 메뉴 입니다")

    def pay(self):
        print(f"총 합계 금액은 {self.price_sum}원 입니다.")
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('현금 결제는 1번,카드 결제는 2번을 입력해 주세요 :'))

            if self.pay_method == 1:
                print("직원을 호출하겠습니다.")
            elif self.pay_method == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
            else:
                print("다시 결제를 시도해 주세요.")

    def table(self):
        outline1 = '⟝' + '-' * 30 + '⟞'
        outline2 = '|' + ' ' * 31 + '|'
        print(outline1)
        for i in range(5):
            print(outline2)

        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])
        print('합계 금액 :', self.price_sum)

        for i in range(5):
            print(outline2)
        print(outline1)


# 4) 출력을 볼 수 있게 명령어 추가
kioskTest = Kiosk()
kioskTest.menu_print()
kioskTest.menu_select()
kioskTest.pay()
kioskTest.table()
```
